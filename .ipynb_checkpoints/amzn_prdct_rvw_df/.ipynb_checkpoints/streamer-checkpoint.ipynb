```python


# Initialize Spark Session (with Hive support if needed, for potential future use)
spark = SparkSession.builder \
    .appName("ReviewStreamingApp") \
    .master("local[*]") \
    .enableHiveSupport() \
    .config("spark.sql.streaming.checkpointLocation", "/tmp/checkpoint/review_stream") \
    .getOrCreate()


# Define schema matching your CSV
review_schema = StructType([
    StructField("product_id", StringType(), True),
    StructField("product_category", StringType(), True),
    StructField("review_rating", DoubleType(), True),
    StructField("review_text", StringType(), True),
    StructField("review_date", DateType(), True),
    StructField("verified_purchase", StringType(), True),
    StructField("price", DoubleType(), True)
])

# 1. Monitor directory /tmp/review_data for incoming CSV files
raw_stream_df = spark \
    .readStream \
    .schema(review_schema) \
    .option("header", "true") \
    .option("maxFilesPerTrigger", 1) \
    .csv("/tmp/review_data")

# 2. Write streaming data to HDFS in Parquet format (partitioned)
hdfs_output_path = "hdfs://localhost:9000/user/reviews/parquet"

write_query = raw_stream_df.writeStream \
    .format("parquet") \
    .option("path", hdfs_output_path) \
    .option("checkpointLocation", "/tmp/checkpoint/write_parquet") \
    .partitionBy("product_category", "review_date") \
    .trigger(processingTime='30 seconds') \
    .outputMode("append") \
    .start()

# 3. Compute continuous aggregations: total reviews & avg rating per product_category
aggregations_df = raw_stream_df \
    .groupBy("product_category") \
    .agg(
        count("*").alias("total_reviews"),
        round(avg("review_rating"), 2).alias("avg_rating")
    ) \
    .orderBy(desc("total_reviews"))

# Print results to console continuously
console_query = aggregations_df.writeStream \
    .outputMode("complete") \
    .format("console") \
    .option("truncate", "false") \
    .trigger(processingTime='30 seconds') \
    .start()

# Keep both streaming queries running
console_query.awaitTermination()
write_query.awaitTermination()
```
